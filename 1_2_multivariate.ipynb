### 4. Для целевой и каждого из признаков постройте парную регрессионную модель. Визуализируйте линию регрессии и исходные данные на графике. При помощи коэффициентов оцените силу связи. 1*СВ

```python
for i in range(x.shape[1]):
    model = LinearRegression()
    model.fit(x[[i]], y)

    plt.scatter(x[i], y,  label='Данные')
    xx = np.linspace(x[i].min(), x[i].max(), 100).reshape((-1, 1))
    plt.plot(xx, model.predict(xx), color='red', label='Регрессия')
    plt.title(f'Парная регрессия для признака {i+1}')
    plt.xlabel(f'Признак {i+1}')
    plt.ylabel('Целевая переменная')
    plt.legend()
    plt.show()

    print(f"Коэффициент детерминации (R^2) для признака {i+1}: {model.score(x[[i]], y)}")
```

### 2. Для целевой переменной и каждого из признаков постройте полиномиальную регрессию второго, третьего и десятого порядков. Сделайте вывод о значимости регрессии.

```python
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import PolynomialFeatures


degrees = [2, 3, 10]
for i in range(x.shape[1]):
    for degree in degrees:
        model = make_pipeline(PolynomialFeatures(degree), LinearRegression())
        model.fit(x[[i]], y)

        plt.scatter(x[i], y, color='#1f77b4', label='Данные')
        xx = np.linspace(x[i].min(), x[i].max(), 100).reshape((-1, 1))
        plt.plot(xx, model.predict(xx), color='red', label=f'Полиномиальная регрессия {degree} порядка')
        plt.title(f'Полиномиальная регрессия для признака {i+1}')
        plt.xlabel(f'Признак {i+1}')
        plt.ylabel('Целевая переменная')
        plt.legend()
        plt.show()

        r2 = model.score(x[[i]], y)
        print(f"Коэффициент детерминации (R^2) для признака {i+1} и полиномиальной регрессии {degree} порядка: {r2}")
```

### 3. Постройте сводную таблицу показателей точности всех построенных моделей. Сделайте вывод, какие модели переобученнные, какие - недообученные.

```python
results = []
```

Парные регрессии

```python
for i in range(x.shape[1]):
    model = LinearRegression()
    model.fit(x[[i]], y)
    r2 = model.score(x[[i]], y)
    results.append(('Парная регрессия', f'Признак {i+1}', r2))
```

Множественная регрессия

```python
multiple_model = LinearRegression()
multiple_model.fit(x, y)
r2 = multiple_model.score(x, y)
results.append(('Множественная регрессия', 'Все признаки', r2))
```

Полиномиальные регрессии

```python
for i in range(x.shape[1]):
    for degree in degrees:
        model = make_pipeline(PolynomialFeatures(degree), LinearRegression())
        model.fit(x[[i]], y)
        r2 = model.score(x[[i]], y)
        results.append((f'Полиномиальная регрессия {degree} порядка', f'Признак {i+1}', r2))
```

```python
df_r = pd.DataFrame(results, columns=['Тип модели', 'Признак', 'R^2'])
```

### **1. Как должны выглядеть данные для модели машинного обучения?**

- **Табличный формат:**  
  Данные обычно представляют собой таблицу, где:
  - **Строки** — объекты (наблюдения).
  - **Столбцы** — признаки (фичи) + целевая переменная (если обучаем с учителем).

- **Примеры:**  
  | Площадь (м²) | Этаж | Цена (₽) |
  |--------------|------|-----------|
  | 50           | 3    | 6 000 000 |
  | 70           | 5    | 8 000 000 |

- **Важно:**
  - Признаки желательно нормализовать.
  - Пропуски и выбросы — нужно обработать.
  - Категориальные данные — закодировать (OneHot, LabelEncoder и т.п.).

---

### **2. В чем физический смысл коэффициентов линейной регрессии?**

- **Коэффициенты линейной регрессии** показывают, **насколько сильно и в какую сторону влияет каждый признак на результат**.

- Пример:
  \[
  \text{Цена} = 100\,000 \times \text{Площадь} + 200\,000 \times \text{Этаж} + 1\,000\,000
  \]
  → Площадь увеличивается на 1 м² → цена вырастет на 100 000 ₽.

---

### **3. Чем множественная регрессия отличается от парной?**

- **Парная регрессия:** один признак → одна целевая переменная.  
  \[
  y = a \cdot x + b
  \]

- **Множественная регрессия:** несколько признаков → одна целевая.  
  \[
  y = a_1x_1 + a_2x_2 + \ldots + a_nx_n + b
  \]

Множественная — обобщение парной и используется в большинстве реальных задач.

---

### **4. Какую функцию ошибки использует модель линейной регрессии?**

- **Mean Squared Error (MSE)** — наиболее распространённая:
  \[
  MSE = \frac{1}{n} \sum (y_{\text{true}} - y_{\text{pred}})^2
  \]
- Минимизация этой функции позволяет находить **наилучшие параметры модели**.

---

### **5. Какие метрики эффективности можно применять для оценки регрессионных моделей?**

- **MSE** — средняя квадратичная ошибка.
- **RMSE** — корень из MSE (интерпретируется в тех же единицах, что и целевая переменная).
- **MAE** — средняя абсолютная ошибка.
- **R² (коэффициент детерминации):**
  - От 0 до 1 (или даже отрицательное значение).
  - Показывает, **сколько дисперсии объясняет модель**.
  - R² = 1 — идеальное предсказание.

---

### **6. Какие еще регрессоры есть в библиотеке `sklearn`?**

- `LinearRegression` — простая линейная модель.
- `Ridge` — линейная модель с L2-регуляризацией.
- `Lasso` — линейная модель с L1-регуляризацией.
- `ElasticNet` — комбинация L1 и L2.
- `DecisionTreeRegressor` — дерево решений.
- `RandomForestRegressor` — ансамбль деревьев.
- `SVR` — метод опорных векторов.
- `KNeighborsRegressor` — метод k-ближайших соседей.
- `GradientBoostingRegressor` — градиентный бустинг.

---

### **7. Почему для визуализации модели нельзя использовать предсказанные значения по обучающему датасету?**

- Это может создать **ложное впечатление, что модель работает идеально** — ведь она "видела" эти данные.
- **Правильнее** — визуализировать модель **на валидационной или тестовой выборке**, чтобы увидеть, **как она обобщает**.
- Особенно важно, если мы хотим понять, **не переобучилась ли модель**.
